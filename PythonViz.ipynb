```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
import requests
import pandas as pd
```

```python
df = pd .read_csv("/content/sample_data/Top_Highest_Openings.csv")
```

```python
print(df)
```

```python
values = list(df.groupby('Distributor')['Total Gross'].mean())
labels = list(set(list(df['Distributor'])))
print(max(values))
```

```python
fig=plt.figure()
ax=plt.axes()
plt.xticks(rotation=90)
ax.bar(labels,values)
ax.set_ylabel("Total Gross in Hundreds of Millions of Dollars")
ax.set_xlabel("Distributor")
ax.set_title('Mean Total Gross by Distributor')
```

```python
x=list(df['Theaters'])
y=list(df['Average'])
fig,ax=plt.subplots()
ax.scatter(x,y,s=80,marker=(5,0),color='purple')
ax.set_ylabel("Average Revenue for each Theater where Movie Played")
ax.set_xlabel("Number of Theaters where Movie Played")
ax.set_title('Theater Revenue')
```

```python
fig,ax=plt.subplots()
ax.boxplot(list(df['% of Total']),1)
ax.set_ylabel("% Of Total Revenue Earned in a Movie's Opening Weekend")
ax.set_title("Opening Weekend Revenue")
```
